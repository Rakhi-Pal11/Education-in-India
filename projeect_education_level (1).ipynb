# ANALYSIS BASED ON TOTAL NUMBER OF PERSON

```python
#taking the number of total peaople only for all ages and where area is mention as total
filtered_df = data[(data['Age-group'] == 'All ages') & (data['Total/Rural/Urban'] == 'Total')]
```

# COMPARE TOTAL PERSON WITH LITERATE PERSON

```python
#taking the number of total peaople only for all ages and where area is mention as total
filtered_df = data[(data['Age-group'] == 'All ages') & (data['Total/Rural/Urban'] == 'Total')]


# Group the data by 'Area Name' and sum up the 'Total Persons' and 'Literate Persons'
area_wise_data =filtered_df.groupby('Area Name').agg({
    'Total Persons': 'sum',
    'Literate Persons': 'sum'
})

# Plotting the data
plt.figure(figsize=(12, 8))
bar_width = 0.35
index = np.arange(len(area_wise_data))

# Bar chart for Total Persons
plt.bar(index, area_wise_data['Total Persons'], bar_width, label='Total Persons', color='skyblue')

# Bar chart for Literate Persons
plt.bar(index + bar_width, area_wise_data['Literate Persons'], bar_width, label='Literate Persons', color='lightgreen')

# Adding titles and labels
plt.title('Comparison of Total Persons and Literate Persons by Area')
plt.xlabel('Area Name')
plt.ylabel('Number of Persons')
plt.xticks(index + bar_width / 2, area_wise_data.index, rotation=90)
plt.legend()

# Adjust layout to fit x-axis labels
plt.tight_layout()

# Show the plot
plt.show()

# Save the plot as an image file
plt.savefig('area_wise_comparison_total_vs_literate.png')
```

CONCLUSION

```python
#by this way we can answer some similer question like
#1>area wise comparison of total person with total illiterate person
#2>area wise comparison of total person with total males
#3>area wise comparison of total person with total females
#4>area wise comparison of literate person with total males
#5>area wise comparison of literate person with total females
#7>area wise comparison of  total females with total literate females
#7>area wise comparison of  total males with total literate males
#8>area wise comparison of  total females with total illiterate females
#9>area wise comparison of  total males with total illiterate males
#ETC
```

# AREA TYPE WISE ANALYSIS

```python
# Bar Plot: Total number of people at different educational levels across area types
#taking the number of total peaople only for all ages 
filtered_df = data[data['Age-group'] == 'All ages']
# Sum the data by 'Total/Rural/Urban'
education_totals = filtered_df.groupby('Total/Rural/Urban')[
    [
        'Illiterate Persons',
        'Literate Persons',
        'Below primary Persons',
        'Primary level Persons',
        'Middle Persons',
        'Matric/Secondary Persons',
        'Higher secondary Persons',
        'Non-technical diploma Persons',
        'Technical diploma Persons',
        'Graduate & above Persons',
        'Unclassified Persons'
    ]
].sum().T

# Plot the bar chart
education_totals.plot(kind='bar', figsize=(14, 8))
plt.title('Total Number of People at Different Educational Levels Across Area Types')
plt.xlabel('Educational Levels')
plt.ylabel('Number of People')
plt.xticks(rotation=45)
plt.legend(title='Area Type', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
```

```python
# Pie Chart: Distribution of Graduate & above Persons by Area Type (Total, Rural, Urban)
#taking the number of total peaople only for all ages 
filtered_df = data[data['Age-group'] == 'All ages']
# Sum the data by 'Total/Rural/Urban'
area_distribution = filtered_df.groupby('Total/Rural/Urban')['Graduate & above Persons'].sum()

# Plot the pie chart
plt.figure(figsize=(8, 8))
plt.pie(area_distribution, labels=area_distribution.index, autopct='%1.1f%%', startangle=140, colors=sns.color_palette("pastel"))
plt.title('Distribution of Graduate & above Persons by Area Type')
plt.show()
```

CONCLUSION

```python
#by this analysis we can say rural area has more graduate and above person then urban area.
#using this we can give answer of some similer question like-
#1>Area Type (Total, Rural, Urban) wise distribution of total person,total literate total illiterate 
#2>Area Type (Total, Rural, Urban) wise distribution of total primary level ,below primary etc.
```

# COMPARE TECHNICAL AND NON TECHNICAL DEPLOMATIC PERSON

```python
# Visualization 3: Stacked Bar Plot - Technical vs Non-Technical Diploma Holders by State
#taking the number of total peaople only for all ages and where area is mention as total
filtered_df = data[(data['Age-group'] == 'All ages') & (data['Total/Rural/Urban'] == 'Total')]
numeric_columns = filtered_df.select_dtypes(include=[np.number]).columns

statewise_data = filtered_df.groupby('Area Name')[numeric_columns].sum()


plt.figure(figsize=(14,8))
statewise_data[['Technical diploma Persons', 'Non-technical diploma Persons']].plot(kind='bar', stacked=True, colormap='Paired', figsize=(14,8))
plt.title('Technical vs Non-Technical Diploma Holders by State/UT', fontsize=16)
plt.xlabel('State', fontsize=12)
plt.ylabel('Number of Diploma Holders', fontsize=12)
plt.xticks(rotation=90)
plt.show()
```

conclusion

```python
#by this analysis we can see that for mostly all the state the number of technical deploma is greater than non technical deploma
#using this way we can answer some similer question like -
#1>area wise distribution of technical deplomatic males and females
```

DISTRIBUTION OF TOTAL PERSON ACROSS AGE GROUP

```python
filtered_df = data[(data['Age-group'] != 'All ages') & (data['Total/Rural/Urban'] == 'Total')]
# Group by 'Age-group' and sum the 'Total Persons'
age_group_data =filtered_df.groupby('Age-group')['Total Persons'].sum().reset_index()

plt.figure(figsize=(10, 7))
plt.pie(age_group_data['Total Persons'], labels=age_group_data['Age-group'], autopct='%1.1f%%', colors=plt.cm.Paired.colors)
plt.title('Distribution of Total Persons Across Age Groups')
plt.show()
```

PROPORTION OF MALES AND FEMALES IN A SPECIFIC AGE GROUP

```python
filtered_df = data[data['Total/Rural/Urban'] == 'Total']

specific_age_group =filtered_df [filtered_df ['Age-group'] == "10"] # Change the age group as needed
total_males = specific_age_group['Total Males'].values[0]
total_females = specific_age_group['Total Females'].values[0]

plt.figure(figsize=(8, 8))
plt.pie([total_males, total_females], labels=['Males', 'Females'], autopct='%1.1f%%', colors=['lightgreen', 'pink'])
plt.title('Males vs Females in Age Group 10')
plt.show()
```

CONCLUSION

AREA WISE DISTRIBUTION OF DIFFERENT AGE GROUP

```python
filtered_df = data[(data['Age-group'] != 'All ages') & (data['Total/Rural/Urban'] == 'Total')]


# Group the data by 'Area Name' and 'Age-group'
grouped_data =filtered_df.groupby(['Area Name', 'Age-group']).agg({
    'Total Persons': 'sum',
    'Literate Persons': 'sum'
}).reset_index()

# Display the first few rows of the grouped data
print(grouped_data.head())

# Pivot the data for easier visualization
pivot_table = grouped_data.pivot(index='Area Name', columns='Age-group', values='Literate Persons')

# Plotting the data
pivot_table.plot(kind='bar', stacked=True, figsize=(14, 8), colormap='viridis')
plt.title('Total literate Persons by Age Group and Area')
plt.xlabel('Area Name')
plt.ylabel('Total literate Persons')
plt.xticks(rotation=90)
plt.legend(title='Age-group')
plt.tight_layout()

# Show the plot
plt.show()


```

LITERATE PERSON ACROSS DIFFERENT AGE GROUP BY AREA TYPE

```python
filtered_df = data[data['Age-group'] != 'All ages']

grouped_data = filtered_df.groupby(['Total/Rural/Urban', 'Age-group']).sum().reset_index()

# Visualize: Literate Persons Across Different Age Groups by Area Type
plt.figure(figsize=(30, 8))
sns.barplot(data=grouped_data, x='Age-group', y='Literate Persons', hue='Total/Rural/Urban')
plt.title('Literate Persons Across Different Age Groups by Area Type')
plt.xlabel('Age Group')
plt.ylabel('Number of Literate Persons')
plt.legend(title='Area Type')
plt.show()
```

# RELATIONSHIP BETWEEN LITERATE AND GRADUATE ACROSS AGE GROUP

```python
# Scatter Plot: Relationship between Literate Persons and Graduate & above Persons across Age Groups
filtered_df = data[data['Age-group'] != 'All ages']

plt.figure(figsize=(12, 6))
sns.scatterplot(data=filtered_df, x='Literate Persons', y='Graduate & above Persons', hue='Total/Rural/Urban', style='Age-group', s=100)
plt.title('Literate Persons vs. Graduate & above Persons by Area Type and Age Group')
plt.xlabel('Literate Persons')
plt.ylabel('Graduate & above Persons')
plt.legend(title='Area Type and Age Group', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
```

# CORRELATION BETWEEN DIFFERENT EDUCATION LEVEL

```python
filtered_df = data[(data['Age-group'] != 'All ages') & (data['Total/Rural/Urban'] == 'Total')]


# Select relevant columns for correlation
education_levels_data = filtered_df[
    [
        'Illiterate Persons',
        'Literate Persons',
        'Below primary Persons',
        'Primary level Persons',
        'Middle Persons',
        'Matric/Secondary Persons',
        'Higher secondary Persons',
        'Non-technical diploma Persons',
        'Technical diploma Persons',
        'Graduate & above Persons',
        'Unclassified Persons'
    ]
]

# Compute the correlation matrix
corr_matrix = education_levels_data.corr()

# Plot the heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Correlation Matrix of Educational Levels')
plt.show()
```
